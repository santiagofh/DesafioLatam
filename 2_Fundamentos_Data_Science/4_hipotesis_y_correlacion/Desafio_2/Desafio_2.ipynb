- Si el P-VALUE es de un 0,1183 se interpreta que no es suficientemente pequeña para rechazar la hipotesis nula, bajo las condiciones que la probabilidad de la hipotesis nula se distribuye de forma normal y con una confianza del 95%. 
- la máxima confianza con la que podemos rechazar la hipótesis nula (considerando el p-valor obtenido) es de un 88,2%.

```python
import pandas as pd
import numpy as np

df=pd.read_csv('nations.csv', encoding='iso-8859-1')
regiones=df.region.unique()
for i in regiones:
    df[i]=np.where(df.region==i,1,0)
```

```python
#3
def prueba_hip(df,variable,binarize):
    import scipy.stats
    df.dropna(subset=[variable],inplace=True)
    df_pertenece=df.loc[df[binarize]==1,variable]
    df_no_pertenece=df.loc[df[binarize]==0,variable]
    media_pertenece=df_pertenece.mean()
    media_no_pertenece=df_no_pertenece.mean()
    diferencia_media=media_pertenece-media_no_pertenece
    t,pval=scipy.stats.ttest_ind(df_pertenece,df_no_pertenece)
    print(f'grupo 0, media: {media_no_pertenece}')
    print(f'grupo 1, media: {media_pertenece}')
    print(f'diferencia entre las medias: {diferencia_media}')
    print(f'valor Z: {t}')
    print(f'p-valor:{pval}')
```

```python
prueba_hip(df,"chldmort","Africa")
```

```python
prueba_hip(df,"adfert","Africa")
```

```python
prueba_hip(df,"life","Africa")
```

```python
#4
def grafica_hist(df,variable,binarize):
    import seaborn as sns
    import matplotlib.pyplot as plt
    df.dropna(subset=[variable],inplace=True)
    plt.figure(figsize=(10.0,10.0))
    plt.ylabel("Cuenta")
    plt.title(f"Comparación Distribución de {variable} para {binarize} v/s Resto del mundo")
    sns.histplot(df,x=variable,hue=binarize,alpha=0.1,kde=True)
```

# ASIA

```python
prueba_hip(df,"life","Asia")
grafica_hist(df,"life","Asia")
```

- Segun los valores entregados, la diferencia de medias de ambos grupos no difieren considerablemente
- No se rechaza la hipótesis nula de que ambas medias son iguales
