```python
# imports cl치sicos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
# importamos nuestro archivo con funciones auxiliares
import seaborn as sns
plt.style.use('seaborn')
plt.rcParams['figure.figsize']=(8,6)
```

```python
df=pd.read_csv('nations.csv', encoding='iso-8859-1')
```

```python
def plot_mean(data,**kwargs):
    m = data.mean()
    plt.axvline(m, **kwargs)
    return m

def binarize_histogram(dataframe, variable):
    tmp = dataframe
    tmp['binarize'] = np.where(tmp[variable] > np.mean(tmp[variable]), 1,
    0)
    g=sns.FacetGrid(tmp,col_wrap=2,col='binarize',size=6)
    g.map(sns.histplot,variable,alpha=.5)
    g.map(plot_mean,variable,color='blue')

binarize_histogram(df,'adfert')
```

## Interpretacu칩n de los graficos:
- Se observa que existe una mayor condensaci칩n de los datos menor a la media

```python
def grouped_boxplot(dataframe,variable, group_by):
    tmp = dataframe
    g=sns.FacetGrid(tmp,col=group_by,size=3)
    g.map(sns.boxplot,variable,orient="v")
grouped_boxplot(df,'adfert','region')
```

```python
def grouped_scatterplot(df,col,x1,y1):
    g=sns.FacetGrid(df,col=col,size=3)
    g.map(sns.scatterplot,data=df,x=x1, y=y1)
grouped_scatterplot(df,'region','school','adfert')
```
