## Variables Aleatorias

- Continuas: Distribución normal. Un ejemplo de esto es que el indice de desarrollo humano se asemeja a una distribución normal. Depende de dos parametros:
 - Media
 - Varianza
Inspeccionar que tanto se acopla el histograma a la dist. normal
En el parametro media gobernara el punto donde se concentra la mayoria de los datos
El parametro varianza es que tan concentrados estan los datos en torno a la media

```python
mu=hdi_dropna.mean()
sigma=hdi_dropna.var()
std=hdi_dropna.std()
```

```python
print("la media es {} y a desviación es {}".format(mu,std))
```

## Variables Aleatorias 2
Tenemos que acoplar la curva Gausiana


```python
mu=hdi_dropna.mean()
sigma=hdi_dropna.var()
std=hdi_dropna.std()
simulate_normal=np.random.normal(mu,std,1000)
plt.hist(simulate_normal, color='lightblue',alpha=0.4,density=True, stacked=True)
plt.xlim(.3,1)

from scipy.stats import norm
x_axis=np.linspace(.3,1,1000)
simiulate_normal_pdf=norm.pdf(x_axis,mu,std)
plt.plot(x_axis, simiulate_normal_pdf,color='tomato', lw=3)
```

Lo que tenemos aca es como se debe comportar de forma teorica el idh

```python
from scipy.stats import gaussian_kde
plt.hist(hdi_dropna, color='lightgreen', alpha=.4, density=True, stacked=True)
x_axis=np.linspace(.3,1,100)
gauss_norm=gaussian_kde(hdi_dropna)
z=np.reshape(gauss_norm(x_axis).T,x_axis.shape)
plt.plot(x_axis,z,color='tomato',lw=3)
```

Vamos a hacer una transformación logaritmica

## Distribución normal estandarizada (media=0, varianza=1)


```python
gfx.normal_distribution_sigma()
```

Elementos de una dis. normal estandarizada
- Es unimodal, tiene un punto mas alto de datos
- Es simetrica, los datos gravitan dentro de la media
- Es positiva, los valores en el rango tienen una prob de ocurrencia


Mediante ambos componentes: Media y varianza podemos calcular el porcentaje de datos que estan bajo la curva.
- 68% se encuentra entre -1 y 1 std, la mayoria de los datos
- 2 std se encuentra el 95% de los datos
- 3 std se encuentra el 99% de los datos


## Puntajes z 
La información que entrega la dist. normal. Se puede utilizar como una obs se comporta respecto a la media.
- Obtener la media del vector
- Restar la obs especifica respecto a la media
- Resultado / std 
Esto nos indica a cuantas std se encuentra un valor respecto a la media 

```python
chile_hdi=df.loc[df.cname=='Chile',['undp_hdi']]
chile_hdi=df[df['cname']=='Chile']['undp_hdi']
chile_zscore=(chile_hdi - df['undp_hdi'].mean())/df['undp_hdi'].std()
```

El caso de Chile se encuentra 0.96 std sobre la media. Se encuentra superior a la media del mundo en indice de desarrollo humano

```python
def get_cscore(columna,caso):
    get_point=df[df['cname']==caso][columna]
    preproc=df[columna].dropna()
    get_mean=preproc.mean()
    get_std=preproc.std()
    return float(get_point-get_mean)/get_std
```

```python
chile_zscore
```

```python
get_cscore('undp_hdi','Chile')
```

```python
def point_zscore(columna,caso):
    get_point=df[df['cname']==caso][columna]
    preproc=df[columna].dropna()
    get_mean=preproc.mean()
    get_std=preproc.std()
    get_zscore = float(get_point-get_mean)/get_std
    plt.axvline(get_zscore, ymin=0,ymax=.1,color='tomato',lw=2)
    plt.annotate(caso,xy=(get_zscore + .05, .06), color='tomato',rotation=60,size=7,ha='center')
```

```python
import numpy as np

data = np.random.normal(loc=10, scale=2, size=1000)
print(np.mean(data), np.std(data))
import matplotlib.pyplot as plt
plt.hist(data)
from scipy.stats import norm
dist = norm(loc=np.mean(data), scale=np.std(data))
x = np.linspace(dist.ppf(0.001),
                dist.ppf(0.999), 100)
plt.hist(data, normed=True, label="Muestra")
plt.plot(x, dist.pdf(x), "r-", label="PDF teórico")
plt.legend()
```
